# Bivariate Plot

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
sns.displot(x='total_bill',y='tip', data=tips)
plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.displot(x='total_bill', y='tip', data=tips, rug=True,
kind='hist', bins=30)
plt.show()
```

```python
# A more interesting type of visualization is a bivariate KDE plot.
```

```python
tips = sns.load_dataset("tips")
sns.displot(x='total_bill', y='tip', data=tips, kind='kde')
plt.show()
```

```python
# You can add a rug plot to the output as follows
```

```python
tips = sns.load_dataset("tips")
sns.displot(x='total_bill', y='tip',data=tips, rug=True,kind='kde')
plt.show()
```

```python
# Based on the columns in the dataframe, you can create individual visualizations
# arranged in rows or columns. Let’s create a visualization based on the size of tips as
# follows
```

```python
sns.displot(x='total_bill', y='tip',data=tips, rug=True,kind='kde', col='size')
plt.show()
```
